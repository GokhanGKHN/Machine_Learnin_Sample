```python
#Veri Görselleştirme
import matplotlib.pyplot as plt
import seaborn as sns
#Hesaplamalar 
import numpy as np
import pandas as pd
```

```python
veriSeti=pd.read_csv("insurance.csv") #veri setini okuma
```

```python
veriSeti=veriSeti.rename(columns={  #kolon adlarını değiştirme 
                                    "age":"yas",
                                    "sex":"cinsiyet",
                                    "bmi":"vki",
                                    "children":"cocukSayisi",
                                    "smoker":"sigaraDurum",
                                    "region":"bolge",
                                    "charges":"odemeMiktar"})
```

```python
veriSeti["cinsiyet"] =veriSeti["cinsiyet"].replace( #nitelik adlarını değiştirme
                                                    ["male","famale"],["erkek","kadın"]
                                                        )
veriSeti["sigaraDurum"] =veriSeti["sigaraDurum"].replace( #nitelik adlarını değiştirme
                                                    ["no","yes"],["hayir","evet"]
                                                        )
veriSeti["bolge"] =veriSeti["bolge"].replace( #nitelik adlarını değiştirme
                                                    ["southeast","northwest","southwest","northeast"],["guneydogu","kuzeybati","guneybati","kuzeydogu"]
                                                        )
```

```python
veriSeti.head(100) #ilk üc satırı getir. 
```

```python
veriSeti.dtypes #veri tiplerini görüntüleme 
```

```python
veriSeti["cinsiyet"]=veriSeti["cinsiyet"].astype('category') #Veri Tipi değiştirme object > category
veriSeti["sigaraDurum"]=veriSeti["sigaraDurum"].astype('category')
veriSeti["bolge"]=veriSeti["bolge"].astype('category')
```

```python
veriSeti.describe()#tablo özeti (numeric)
```

```python

pd.set_option("display.max_columns",20) #yan yana gösterilecek kolon sayısı 
veriSeti.describe(include="all") #tablo özeti (hepsi)
```

#  Sütun Grafiği (bar chart) 

```python
ozet=veriSeti.groupby("sigaraDurum")["odemeMiktar"].mean()
plt.bar(x=ozet.index,height=ozet.values,color="r")
plt.xlabel("Sigara İçme Durumu")
plt.ylabel("Ödeme Miktarı")
```

```python
def etiketEkle(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha="center")

plt.bar(x=ozet.index,height=ozet.values,color="r")
plt.xlabel("Sigara İçme Durumu")
plt.ylabel("Ödeme Miktarı")
etiketEkle(ozet.index,ozet.values.round(2))
```

```python
plt.bar(x=ozet.index,height=ozet.values,color="r",width=0.95)
#plt.barh(y=ozet.index,width=ozet.values,color="b",height=0.2)
```

# pasta grafiği (pie chart)
Frekans ve yüzde değerlernden yola çıkılarak kategork bir nitelk pasta grafiğ ile kolaylıkla ifade edilebilr.

```python
ozet=veriSeti.groupby("bolge")["odemeMiktar"].sum()
etiketler=ozet.index
degerler=ozet.values
renkler=sns.color_palette("viridis",4)
secim=(0,0,0.2,0)
plt.pie(degerler,explode=secim,labels=etiketler,autopct="%%%4.1f",shadow=True,startangle=360,colors=renkler)
```

```python
sns.scatterplot(x="yas",y="odemeMiktar",hue="sigaraDurum",data=veriSeti) #Grafiğe 2.boyut katmak. 
```

```python
#Girafiğe üçüncü boyut katma
fig=plt.figure(figsize=(10,6))
ax=plt.axes(projection="3d")
ax.scatter(veriSeti.yas,veriSeti.odemeMiktar,veriSeti.vki,color="darkblue")
ax.set_xlabel("Yaş")
ax.set_ylabel("Ödeme Miktarı")
ax.set_zlabel("Vucut Kitle İndeksi",rotation=90)
ax.zaxis.labelpad=-0.7
plt.title("3-Boyutlu Serpilme Diyagramı")
```

# Kutu-bıyık grafiği (box and whisker plot)
kutu grafğ brden fazla grubun dağılımını brbryle karşılaştırmak çn sıklıkla kullanılan grafk türlernden brdr

```python
veriSeti["vki"].describe()
```

```python
bp=plt.boxplot(veriSeti["vki"], 
               vert=True, #Grafik dikey çizilecek
               showmeans=True,#Ortalama ve ortalama cizgisinin gösterileceği parametreler
               meanline=True, 
               labels=("x"),
               patch_artist=True,
               medianprops={'linewidth':2,'color':'purple'}, #Ortanca değer ve ortalama cizgilerine ait özellikleri gösterir. 
               meanprops={'linewidth':2,'color':'red'},
               notch=True #Çentik eklemek (güven aralığını gösterir ve default %95)
               )
```

```python
sns.boxplot(y="sigaraDurum",
            x="odemeMiktar",
            data=veriSeti,
            palette="rainbow")
```
